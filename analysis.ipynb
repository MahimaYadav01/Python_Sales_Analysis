```python
#BAR CHART FOR GENDER AND IT'S COUNT

sf=sns.countplot(x='Gender',data=df,hue='Gender')
for bars in sf.containers:
    sf.bar_label(bars)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
#BAR CHART BETWEEN GENDER VS AMOUNT

sales=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
gh=sns.barplot(x='Gender',y='Amount',data=sales,hue='Gender', palette={'F': 'lightcoral', 'M': 'steelblue'},legend=True)
for bars in gh.containers:
    gh.bar_label(bars)
sns.set(rc={'figure.figsize':(5,5)})   
```

```python
#TOTAL AMOUNT
df['Amount'].sum()
```

```python
#TOTAL AMOUNT SPENT BY EACH GENDER
df.groupby('Gender')['Amount'].sum().sort_values(ascending=False)
```

```python
df.columns
```

```python
#BAR CHART FOR FINDING BEST CUSTOMER BASED ON THE AMOUNT SPENT
best_cust=df.groupby(['Cust_name'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sh=sns.barplot(x='Cust_name',y='Amount',data=best_cust,hue='Cust_name')
for bars in sh.containers:
    sh.bar_label(bars)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
# BAR CHART FOR AGE GROUP AND THIER ORDERS
age=df.groupby(['Age Group'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
qf=sns.barplot(x='Age Group', y='Orders',data=age,hue='Age Group')
for bars in qf.containers:
    qf.bar_label(bars)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
#BAR CHART FOR AGE GROUP AND THE AMOUNT THEY HAVE SPENT
age_group_amount=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
jk=sns.barplot(x='Age Group',y='Amount',data=age_group_amount,hue='Age Group')
for bars in jk.containers:
    jk.bar_label(bars)
sns.set(rc={'figure.figsize':(13,5)})
```

```python
#BAR CHART OF AGE GROUP VS GENDER
kl=sns.countplot(x='Age Group' ,data=df,hue='Gender')
for bars in kl.containers:
    kl.bar_label(bars)
sns.set(rc={'figure.figsize':(10,5)})
```

```python
#BAR CHART OF Marital_Status
msb=sns.countplot(data = df, x = 'Marital_Status')
for bars in msb.containers:
    msb.bar_label(bars)
sns.set(rc={'figure.figsize':(5,5)}) 
```

```python
#BAR CHART OF GENDER AND THE AMOUNT SPENT BASED ON MARITAL STATUS
ms = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
msb=sns.barplot(data = ms, x = 'Marital_Status',y= 'Amount', hue='Gender')

for bars in msb.containers:
    msb.bar_label(bars)
sns.set(rc={'figure.figsize':(10,7)})    
```

```python
# total number of orders from top 7 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(7)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders',hue='State')
```

```python
# top 7 states based on the amount they spent
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)

sns.set(rc={'figure.figsize':(15,5)})
msb=sns.barplot(data = sales_state, x = 'State',y= 'Amount',hue='State')
for bars in msb.containers:
    msb.bar_label(bars)
```

```python
#BAR CHART BASED ON COUNT OF ORDERS BY AN OCCUPATION
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#BAR CHART BASED ON AMOUNT SPENT BY AN OCCUPATION
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',hue='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#BAR CHART BASED ON PRODUCT CATEGORY AND THE COUNT OF ORDERS
sd=sns.countplot(x='Product_Category',data=df,hue='Product_Category')
sns.set(rc={'figure.figsize':(15,5)})
for bars in sd.containers:
    sd.bar_label(bars)
```

```python
#BAR CHART BASED ON PRODUCT CATEGORY AND THE AMOUNT EARNED BY IT
pc=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sd=sns.barplot(x='Product_Category',y='Amount',data=pc,hue='Product_Category')
sns.set(rc={'figure.figsize':(15,5)})
for bars in sd.containers:
    sd.bar_label(bars)
```

```python
#BAR CHART BASED ON PRODUCT ID AND ORDERS 
pc=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(7)
sd=sns.barplot(x='Product_ID',y='Orders',data=pc,hue='Product_ID')
sns.set(rc={'figure.figsize':(15,5)})
for bars in sd.containers:
    sd.bar_label(bars)
```

```python
#BAR CHART BASED ON PRODUCT ID AND THE AMOUNT EARNED BY IT
pc=df.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sd=sns.barplot(x='Product_ID',y='Amount',data=pc,hue='Product_ID')
sns.set(rc={'figure.figsize':(15,5)})
for bars in sd.containers:
    sd.bar_label(bars)
```
